```python
x=profit_data.iloc[:,:-1]
y=profit_data.iloc[:,-1]
x
```

```python
miss_col=profit_data[['Youtube Spend','Instagram','Facebook']]
missclm=miss_col.replace(0,np.nan)
missclm.head()
```

```python
profit_d=x.drop(['Youtube Spend','Instagram','Facebook'],axis=1)
x=pd.concat([profit_d,missclm], axis=1)
x.head()
```

```python
x.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lbl=LabelEncoder()
x['Place']=lbl.fit_transform(x['Place'])
X=pd.DataFrame(x)
X.head()
```

```python
pla=pd.get_dummies(X['Place'],drop_first=True)
pla.head()
```

```python
X=pd.concat([X,pla],axis=1)
X.head()
```

```python
X.drop('Place',axis=1,inplace=True)
X.head()
```

```python
X.describe()
```

```python
median=x[['Youtube Spend','Facebook']].median()
X=X.fillna(median)
X
```

```python
median
```

```python
X.isnull().sum()
```

```python
X.dtypes
```

```python
from sklearn.preprocessing import StandardScaler
ytbsclr=StandardScaler()
X['Youtube Spend']=ytbsclr.fit_transform(X[['Youtube Spend']])
X
```

```python
from sklearn.preprocessing import StandardScaler
inssclr=StandardScaler()
X['Instagram']=inssclr.fit_transform(X[['Instagram']])
X
```

```python
from sklearn.preprocessing import StandardScaler
fbsclr=StandardScaler()
X['Facebook']=fbsclr.fit_transform(X[['Facebook']])
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regres=LinearRegression()
regres.fit(X_train,y_train)
```

```python
y_pred=regres.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
